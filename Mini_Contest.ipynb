# Preprocessing

## Import library

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Import dataset

```
dataset = pd.read_csv('data.csv')
dataset.head()
```

## Hapus variabel yang tidak digunakan

```
# Akan dihilangkan variabel "Names", "Onboard_date", "Location", dan "Company"
dataset = dataset.drop(labels=["Names", "Onboard_date", "Location", "Company"], axis=1)
dataset
```

## Definisikan variabel

```
X = dataset.iloc[:, :5].values
X
```

```
Y = dataset.iloc[:, 5].values
Y
```

## Cek data

```
dataset.describe()

# 1. Diperoleh bahwa tidak ada data yang missing (lihat baris count)
# 2. Diperoleh bahwa kolom Total_Purchase mempunyai standar deviasi yang besar, sehingga nantinya akan dilakukan scaling (selain kolom yang dikodekan)
```

```
print(dataset.groupby('Account_Manager').size()) # Variabel "Account_Manager" adalah data kategorik
```

## Bagi data menjadi training set dan test set

```
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```
print(X_train)
```

```
print(X_test)
```

## Scaling

```
# Akan dilakukan scaling pada semua variabel X kecuali "Account_Manager"
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:, :2] = sc.fit_transform(X_train[:, :2])
X_train[:, 3:] = sc.fit_transform(X_train[:, 3:])
X_test[:, :2] = sc.fit_transform(X_test[:, :2])
X_test[:, 3:] = sc.fit_transform(X_test[:, 3:])
```

```
print(X_train)
```

```
print(X_test)
```

# Training dan Test

## Proses training

```
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, Y_train)
```

## Coba prediksi hasil dari X_test

```
Y_pred = classifier.predict(X_test)
```

```
print(Y_pred)
```

## Bandingkan Y_pred dengan Y_test (Y_asli)

```
from sklearn.metrics import confusion_matrix
print(confusion_matrix(Y_test, Y_pred))

# Yang memprediksi benar/tepat adalah 149+19
```

```
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_pred, Y_test))

# Skor akurasi model adalah sebesar 89%
```
